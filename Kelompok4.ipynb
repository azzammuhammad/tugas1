```python
import matplotlib.pyplot as plt

# Data
jenis_makanan = ['Pizza', 'Burger', 'Pasta', 'Salad', 'Steak', 'Sushi', 'Chicken Wings', 'Tacos', 'Nasi Goreng', 'Kebab', 'Fish and Chips', 'Ramen', 'Dim Sum', 'Satay', 'Sate Padang']
jumlah_pelanggan = [30, 25, 20, 15, 12, 10, 8, 7, 6, 5, 4, 3, 3, 2, 1]

# Plot
plt.figure(figsize=(10, 8))
plt.pie(jumlah_pelanggan, labels=jenis_makanan, autopct='%1.1f%%', startangle=140)
plt.title('Preferensi Makanan Pelanggan Restoran')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Tampilkan chart
plt.show()
```

Chart pie dipilih sebagai teknik visualisasi data karena cocok untuk menunjukkan proporsi atau persentase dari beberapa kategori yang berbeda. Dalam studi kasus ini, kita memiliki beberapa jenis makanan yang menjadi kategori, dan kita ingin menunjukkan seberapa banyak pelanggan yang memilih setiap jenis makanan tersebut.

Berikut adalah alasan-alasan pemilihan teknik visualisasi chart pie:

Menunjukkan Proporsi: Chart pie secara intuitif menunjukkan proporsi atau persentase setiap kategori dalam satu kesatuan (dalam hal ini, keseluruhan preferensi makanan). Ini memungkinkan pembaca untuk dengan cepat melihat bagaimana setiap jenis makanan berkontribusi terhadap total preferensi makanan.
Mudah Dipahami: Chart pie adalah salah satu jenis visualisasi data yang paling mudah dipahami oleh banyak orang. Orang dapat dengan cepat mengidentifikasi bagian mana yang lebih besar atau lebih kecil, dan tidak memerlukan interpretasi yang rumit.
Ringkas: Dalam kasus ini, kami memiliki 15 kategori makanan. Dengan menggunakan chart pie, kita dapat menampilkan semua kategori ini dalam satu visualisasi tanpa membuatnya terlihat terlalu padat atau rumit.
Menciptakan Fokus: Dengan fokus pada bagian-bagian dari pie yang lebih besar (yang menunjukkan preferensi yang lebih tinggi), chart pie membantu dalam menyoroti elemen-elemen yang paling signifikan dalam data. Ini dapat membantu dalam pengambilan keputusan atau penafsiran data yang lebih baik.
Dengan demikian, chart pie adalah pilihan yang tepat untuk menampilkan preferensi makanan pelanggan restoran karena kemampuannya untuk secara jelas menunjukkan proporsi dari masing-masing jenis makanan dan membuatnya mudah dipahami oleh pembaca.

```python
import matplotlib.pyplot as plt

# Data
jenis_produk = ['Elektronik', 'Fashion', 'Perabotan', 'Mainan', 'Makanan', 'Kosmetik', 'Buku', 'Alat Olahraga', 'Perhiasan']
total_penjualan = [120, 90, 80, 60, 110, 70, 45, 55, 65]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(jenis_produk, total_penjualan, color='skyblue')

# Menambahkan judul dan label
plt.title('Pertumbuhan Penjualan E-commerce Berdasarkan Jenis Produk')
plt.xlabel('Jenis Produk')
plt.ylabel('Total Penjualan (dalam ribu USD)')

# Menampilkan label untuk setiap bar
for i in range(len(jenis_produk)):
    plt.text(i, total_penjualan[i] + 2, str(total_penjualan[i]), ha='center')

# Menampilkan plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Pemilihan teknik visualisasi data menggunakan grafik batang dalam studi kasus ini didasarkan pada beberapa pertimbangan:

Jumlah Kategori: Dalam data kita, terdapat beberapa kategori (jenis produk) yang ingin kita bandingkan. Grafik batang sangat cocok untuk menampilkan perbandingan antara beberapa kategori atau kelompok data.
Variabel Kuantitatif: Data kita memiliki variabel kuantitatif (total penjualan) yang ingin kita bandingkan antar kategori. Grafik batang memungkinkan kita untuk dengan jelas memperlihatkan nilai-nilai kuantitatif ini.
Kemudahan Interpretasi: Grafik batang sangat mudah dipahami dan diinterpretasikan oleh banyak orang. Posisi tinggi batang merepresentasikan nilai yang lebih tinggi, sedangkan posisi rendah merepresentasikan nilai yang lebih rendah.
Kemampuan Membandingkan: Grafik batang memungkinkan kita untuk dengan mudah membandingkan nilai antar kategori. Dengan melihat tinggi batang secara relatif, kita dapat dengan cepat menentukan kategori mana yang memiliki total penjualan tertinggi atau terendah.
Menyediakan Informasi Tambahan: Dengan menambahkan label nilai di atas setiap batang, kita dapat memberikan informasi tambahan kepada pembaca tanpa harus melihat skala sumbu. Hal ini membuat grafik menjadi lebih informatif dan mudah dipahami.
Dengan mempertimbangkan semua faktor ini, grafik batang menjadi pilihan visualisasi yang tepat untuk memperlihatkan pertumbuhan penjualan e-commerce berdasarkan jenis produk dalam studi kasus ini.

```python
import matplotlib.pyplot as plt

# Data
prodi = ['S2 Ilmu Komputer', 'S1 Sistem Informasi', 'S1 Informatika', 'S1 Sains Data', 'S1 Manajemen', 'S1 Bisnis Digital', 'D3 Perhotelan', 'S1 Sastra Inggris', 'S1 Akuntansi', 'S1 Multimedia']
jumlah_mahasiswa = [40, 36, 38, 15, 43, 23, 21, 30, 50, 20]

# Membuat bar chart
plt.figure(figsize=(10,6))
plt.bar(prodi, jumlah_mahasiswa, color='skyblue')
plt.xlabel('Jumlah Mahasiswa')
plt.ylabel('Program Studi')
plt.title('Penerimaan Mahasiswa Universitas Nusa Mandiri Tahun 2023')
plt.xticks(rotation=45, ha='right')
plt.show()
```

  1. kita memilih untuk menggunakan bar chart karena:
  Kemudahan Pemahaman: Mudah dipahami karena panjang batang mewakili nilai dan sumbu horizontalnya mewakili kategori.

  Cocok untuk Data Kategori: Cocok untuk membandingkan jumlah antara kategori yang berbeda, seperti penjualan produk.

  Fleksibel: Dapat dimodifikasi dalam berbagai format, termasuk vertikal atau horizontal, serta warna dan gaya batang.

  Visualisasi Data Besar: Dapat menampilkan perbandingan antara banyak kategori tanpa kehilangan jelasnya, cocok untuk data besar.

  Kemampuan Labeling: Memungkinkan penambahan label pada sumbu x dan y, menyertakan informasi tambahan seperti nama kategori atau nilai numerik.
